```python
import cv2 as cv
import matplotlib.pyplot as plt
from skimage import exposure
```

```python
img = cv.imread('penguenler.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img) #BGR renk kodlamasından RGB ye geçiş
```

```python
gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY) #RGB renk kodlamasından GRAY e geçiş
hist = cv.calcHist([gray], [0], None, [256], [0, 256])

dst = cv.equalizeHist(gray)
histEq= cv.calcHist([dst], [0], None, [256], [0, 256])

matched =exposure.match_histograms(gray, dst, multichannel=True)
histMatched = cv.calcHist([matched], [0], None, [256], [0, 256])
```

```python
plt.plot(hist)
plt.show()
plt.plot(histEq)
plt.show()
plt.plot(histMatched)
plt.show()

plt.imshow(cv.cvtColor(gray, cv.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv.cvtColor(dst, cv.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv.cvtColor(matched, cv.COLOR_BGR2RGB))
plt.show()
```
